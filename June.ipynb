```python
#IF
a = 20
b = 20
if a > b:   #note the indentation
  print('a is greater than b')
```

```python
#ELSE
a = 20
b = 20
if a > b:   #note the indentation
  print('a is greater than b')
else:     #note the indentation
  print('a is less than b')
```

```python
#ELIF
a = 20
b = 20
if a > b:   #note the indentation
  print('a is greater than b')
elif a != b:
  print('a is not equal to b')
elif a == b:
  print('a is equal to b')
else:     #note the indentation
  print('a is less than b')
```

```python
#NESTED IF
x = 100
if x > 100:
  print('x is greater than 100')
  if x > 70:
    print('x is greater than 70')
  elif x > 30:
    print('x is greater than 30')
  else:
    print('x is less')
else:
  print('x is less than 30')
```

*   Panda Series

A Pandas Series is a one-dimensional labeled array that can hold any data type. It is a fundamental data structure provided by the Pandas library, which is widely used for data manipulation and analysis in Python. The Pandas Series is similar to a one-column DataFrame or a NumPy array, but it has additional features that make it powerful for data analysis tasks.

Key characteristics of a Pandas Series:

Homogeneous Data: All elements in a Pandas Series must be of the same data type. This data type is inferred from the elements during Series creation.

Labels/Index: Each element in a Series has an associated label or index. By default, the index is a sequence of integers starting from 0, but it can be customized to use labels or a different numeric index.

Similar to a NumPy Array: Internally, a Pandas Series is built on top of a NumPy array, so it inherits many of its properties and methods. However, the Series provides additional functionality, especially in terms of indexing.

Flexible Operations: Series supports various operations, such as mathematical operations, data filtering, and statistical functions. It also supports element-wise operations, similar to NumPy arrays.

### `Pandas DataFrame`:
A Pandas DataFrame is a two-dimensional, labeled data structure used for storing and manipulating tabular data in Python. It is a key component of the Pandas library and provides a convenient way to work with structured data. The DataFrame is similar to a spreadsheet or a SQL table, where data is organized in rows and columns.

Key characteristics of a Pandas DataFrame:

Tabular Structure: Data in a DataFrame is organized in a tabular format with rows and columns. Each column can have a different data type.

Labeled Axes: Both rows and columns have labels, allowing for easy and intuitive indexing and selection of data.

Homogeneous Columns: Each column in a DataFrame must contain elements of the same data type, but different columns can have different data types.

Similar to a NumPy Array: Internally, a Pandas DataFrame is built on top of NumPy arrays, and it inherits many of their properties and methods. However, DataFrames provide additional functionality, especially in terms of indexing and handling missing data.

Flexibility in Data Sources: DataFrames can be created from various data sources, including dictionaries, NumPy arrays, lists, CSV files, Excel files, SQL databases, and more.

Powerful Operations: DataFrames support a wide range of operations, including data cleaning, manipulation, aggregation, filtering, merging, and statistical analysis.

#What is Data?

###Data is a collection of information gathered by observations, measurements, research or analysis. They may consist of facts, numbers, names, figures or even description of things. Data is organized in the form of graphs, charts or tables. There exist data scientist who does data mining and with the help of that data analyse our world.

#Types of Data

###There are 2 main types of data


*   Qualitative data
*   Quatitative data

#Qualitative or Categorical Data

###Qualitative or Categorical Data is a type of data that cant be measured or counted in the form of numbers.

###These types of data are sorted by category, not by number. Thats why it is also known as Categorical Data. These data consist of audio, images, symbols, or text. The gender of a person, i.e., male, female, or others, is qualitative data.

###Qualitative data tells about the perception of people. This data helps market researchers understand the customers tastes and then design their ideas and strategies accordingly.

###Examples of categorical data include; What language do you speak, Favorite holiday destination, Opinion on something (agree, disagree, or neutral), Colors

###Qualitative data is further divided into two parts

#a) Nominal data:
###Nominal Data is used to label variables without any order or quantitative value. The color of hair can be considered nominal data, as one color cant be compared with another color.

###The name nominal comes from the Latin name nomen, which means name. With the help of nominal data, we cant do any numerical tasks or cant give any order to sort the data. These data dont have any meaningful order; their values are distributed into distinct categories.

###Examples include; Colour of hair (Blonde, red, Brown, Black, etc.), Marital status (Single, Widowed, Married), Nationality (Indian, German, American), Gender (Male, Female, Others), Eye Color (Black, Brown, etc.)

#b) Ordinal data:

###Ordinal data have natural ordering where a number is present in some kind of order by their position on the scale. These data are used for observation like customer satisfaction, happiness, etc., but we cant do any arithmetical tasks on them.

###Ordinal data is qualitative data for which their values have some kind of relative position. These kinds of data can be considered in-between qualitative and quantitative data.

###The ordinal data only shows the sequences and cannot use for statistical analysis. Compared to nominal data, ordinal data have some kind of order that is not present in nominal data.

###Examples include; When companies ask for feedback, experience, or satisfaction on a scale of 1 to 10, Letter grades in the exam (A, B, C, D, etc.), Ranking of people in a competition (First, Second, Third, etc.), Economic Status (High, Medium, and Low), Education Level (Higher, Secondary, Primary)

#Quatitative or Numerical Data

###Quantitative data is a type of data that can be expressed in numerical values, making it countable and including statistical data analysis. These kinds of data are also known as Numerical data.

###It answers the questions like how much, how many, and how often. For example, the price of a phone, the computers ram, the height or weight of a person, etc., falls under quantitative data.

###Quantitative data can be used for statistical manipulation. These data can be represented on a wide variety of graphs and charts, such as bar graphs, histograms, scatter plots, boxplots, pie charts, line graphs, etc.

###Examples of Quantitative Data: Height or weight of a person or object, Room Temperature, Scores and Marks (Ex: 59, 80, 60, etc.), Time.

###Quatitative data is further divided into two parts

#a) Discrete data:
###The term discrete means distinct or separate. The discrete data contain the values that fall under integers or whole numbers. The total number of students in a class is an example of discrete data. These data cant be broken into decimal or fraction values.

###The discrete data are countable and have finite values; their subdivision is not possible. These data are represented mainly by a bar graph, number line, or frequency table.

###Examples of discrete data: Total numbers of students present in a class, Cost of a cell phone, Numbers of employees in a company, The total number of players who participated in a competition, Days in a week

#b) Continous Data:
###Continuous data are in the form of fractional numbers. It can be the version of an android phone, the height of a person, the length of an object, etc. Continuous data represents information that can be divided into smaller levels. The continuous variable can take any value within a range.

###The key difference between discrete and continuous data is that discrete data contains the integer or whole number. Still, continuous data stores the fractional numbers to record different types of data such as temperature, height, width, time, speed, etc.

###Examples of continous data: Height of a person, Speed of a vehicle, Time-taken to finish the work , Wi-Fi Frequency, Market share price

#What is Data Merging?

##Data merging is the process of combining two or more datasets into a single dataset. The resulting dataset contains all the columns from the original datasets, and the rows are matched based on a common column or key.

#Types of Data Merging:

##There are several types of data merging, including:

##1. Inner Join: Returns only the rows that have a match in both datasets.
##2. Left Join: Returns all the rows from the left dataset, and the matching rows from the right dataset. If there is no match, the result will contain null values.
##3. Right Join: Similar to left join, but returns all the rows from the right dataset, and the matching rows from the left dataset.
##4. Full Outer Join: Returns all the rows from both datasets, with null values in the columns where there is no match.

#Best Practices:

##1. Use meaningful column names: Use descriptive and consistent column names to make it easier to merge and join datasets.
##2. Use a common key: Use a common column or key to merge and join datasets.
##3. Handle missing values: Use techniques such as imputation or interpolation to handle missing values in the merged dataset.
##4. Verify the results: Verify the results of the merge and join operations to ensure that they are correct and meaningful.

```python
import pandas as pd
```

```python
#Inner Join
df1 = pd.DataFrame({'id': [1, 2, 3], 'name': ['John', 'Mary', 'David']})
df2 = pd.DataFrame({'id': [1, 2, 4], 'age': [25, 31, 42]})

# Perform inner join
df_merged = pd.merge(df1, df2, on='id')

print(df_merged)
```

```python
#Left Join
df1 = pd.DataFrame({'id': [1, 2, 3], 'name': ['John', 'Mary', 'David']})
df2 = pd.DataFrame({'id': [1, 2, 4], 'age': [25, 31, 42]})

# Perform left join
df_merged = pd.merge(df1, df2, on='id', how='left', indicator=True)

print(df_merged)
```

```python
#Right Join
df1 = pd.DataFrame({'id': [1, 2, 3], 'name': ['John', 'Mary', 'David']})
df2 = pd.DataFrame({'id': [1, 2, 4], 'age': [25, 31, 42]})

# Perform right join
df_merged = pd.merge(df1, df2, on='id', how='right', indicator=True)

print(df_merged)
```

```python
#Outer Join
df1 = pd.DataFrame({'id': [1, 2, 3], 'name': ['John', 'Mary', 'David']})
df2 = pd.DataFrame({'id': [1, 2, 4], 'age': [25, 31, 42], 'name': ['John', 'Mary', 'Omo']})

# Perform outer join
df_merged = pd.merge(df1, df2, on='id', how='outer', indicator=True, suffixes=('_df1', '_df2'))

print(df_merged)
```

###Introduction to Hypothesis Testing

Hypothesis testing is a statistical technique used to make inferences about a population based on a sample of data. It involves testing a null hypothesis against an alternative hypothesis.

Basic Concepts in Hypothesis Testing

1. Null Hypothesis (H0): A statement of no effect or no difference.
2. Alternative Hypothesis (H1): A statement of an effect or a difference.
3. Test Statistic: A statistical measure used to determine the likelihood of the null hypothesis.
4. P-Value: The probability of observing the test statistic under the null hypothesis.
5. Significance Level (): The maximum probability of rejecting the null hypothesis when it is true.

Hypothesis Testing with SciPy

SciPy provides functions for performing various types of hypothesis tests, including:

1. `scipy.stats.ttest_ind()`: Performs an independent samples t-test.
2. `scipy.stats.ttest_rel()`: Performs a paired samples t-test.
3. `scipy.stats.f_oneway()`: Performs a one-way ANOVA test.

```python
import numpy as np
from scipy import stats
```

```python
#Independent Samples T-Test
# Sample 1
data1 = np.array([23, 21, 19, 24, 20])

# Sample 2
data2 = np.array([17, 18, 16, 19, 15])

# Perform independent samples t-test
t_stat, p_val = stats.ttest_ind(data1, data2)

print("T-Statistic:", t_stat)
print("P-Value:", p_val)
```

```python
#Paired Samples T-Test
# Sample 1
data1 = np.array([23, 21, 19, 24, 20])

# Sample 2
data2 = np.array([17, 18, 16, 19, 15])

# Perform paired samples t-test
t_stat, p_val = stats.ttest_rel(data1, data2)

print("T-Statistic:", t_stat)
print("P-Value:", p_val)
```

```python
#One-Way ANOVA
# Sample 1
data1 = np.array([23, 21, 19, 24, 20])

# Sample 2
data2 = np.array([17, 18, 16, 19, 15])

# Sample 3
data3 = np.array([25, 26, 27, 28, 29])

# Perform one-way ANOVA
f_stat, p_val = stats.f_oneway(data1, data2, data3)

print("F-Statistic:", f_stat)
print("P-Value:", p_val)
```

###Sampling
Sampling of data in data analysis refers to the process of selecting a subset of data from a larger population or dataset, with the goal of making inferences or drawing conclusions about the entire population.

Why Sampling is Important:

1. Reducing Data Volume: Sampling helps to reduce the volume of data, making it easier to analyze and process.
2. Increasing Efficiency: Sampling can save time and resources, as it eliminates the need to collect and analyze every single data point.
3. Improving Accuracy: Sampling can help to identify biases and errors in the data, leading to more accurate conclusions.
4. Representing the Population: A well-designed sample can provide a representative snapshot of the larger population, allowing for generalizations and predictions.

Types of Sampling:

1. Random Sampling: Every member of the population has an equal chance of being selected.
2. Stratified Sampling: The population is divided into subgroups, and a random sample is taken from each subgroup.
3. Cluster Sampling: The population is divided into clusters, and a random sample is taken from each cluster.
4. Systematic Sampling: Every nth member of the population is selected, where n is a fixed interval.
5. Convenience Sampling: A sample is taken from a convenient or easily accessible subgroup of the population.
6. Purposive Sampling: A sample is taken based on a specific purpose or criterion, such as selecting experts in a particular field.

Sampling Techniques:

1. Simple Random Sampling: A random sample is taken from the population, without replacement.
2. Stratified Random Sampling: A random sample is taken from each subgroup, without replacement.
3. Cluster Random Sampling: A random sample is taken from each cluster, without replacement.
4. Systematic Random Sampling: Every nth member of the population is selected, without replacement.

Sampling Errors:

1. Bias: Systematic errors that can occur due to the sampling method or data collection process.
2. Variability: Random errors that can occur due to chance or random fluctuations.
3. Non-Response: Errors that can occur when some members of the sample do not respond or participate.

Best Practices for Sampling:

1. Define the Population: Clearly define the population of interest.
2. Determine the Sample Size: Calculate the required sample size to achieve the desired level of precision.
3. Choose a Sampling Method: Select a suitable sampling method, based on the research question and population characteristics.
4. Ensure Randomization: Use randomization techniques to minimize bias and ensure representativeness.
5. Monitor and Adjust: Monitor the sampling process and adjust as needed to ensure the sample is representative and accurate.
